# QA Results Review

Reviews `data/outputs/qa_results.csv` produced by `make qa`. The QA framework
covers completeness, conformance, plausibility, uniqueness, referential
integrity, lineage traceability, and mapping coverage.

```python
import pandas as pd

qa = pd.read_csv("../data/outputs/qa_results.csv")
qa["status"].value_counts()
```

## Checks by quality dimension

```python
qa.groupby(["qa_dimension", "status"]).size().rename("checks")
```

## Non-passing checks

```python
failing = qa[qa["status"] != "pass"]
failing[["qa_rule_id", "cdm_table", "field_name", "status", "pass_rate", "remediation_guidance"]]
```

## Notes

High-severity failing checks block a trusted harmonized release. An empty
table above means every acceptance criterion passed in this run.
