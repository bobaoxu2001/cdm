# Source-to-CDM Mapping Catalog Review

Reviews `data/outputs/source_to_cdm_mapping_catalog.csv` produced by
`make mapping`. Mapping rules are config-driven (`config/mapping_rules.yaml`).

```python
import pandas as pd

catalog = pd.read_csv("../data/outputs/source_to_cdm_mapping_catalog.csv")
print(f"{len(catalog)} source fields catalogued")
catalog["mapping_status"].value_counts()
```

## Mapping coverage by source system

```python
mapped = catalog["mapping_status"].isin(["mapped", "partially_mapped"])
(catalog.assign(is_mapped=mapped)
        .groupby("source_system")["is_mapped"]
        .mean()
        .round(3)
        .rename("mapped_or_partial_share"))
```

## Target CDM tables

```python
(catalog.dropna(subset=["target_cdm_table"])
        .groupby("target_cdm_table")
        .size()
        .sort_values(ascending=False)
        .rename("field_count"))
```

## Notes

Fields with `unmapped` or `review_required` status form the steward backlog.
Mapping coverage is checked automatically by QA rule QA-MAP-001.
